```python
from particula import Particula
import matplotlib.pyplot as plt

# Cria a partícula
p = Particula(x=0, y=0, vx=10, vy=10, massa=1)

# Parâmetros
dt = 0.01
fx = 0
fy = -9.8  # força peso (gravidade)

# Loop de simulação
while p.y >= 0:
    p.newton(fx, fy, dt)

# Plotagem da trajetória 
plt.figure(figsize=(8, 5))
plt.plot(p.l_x, p.l_y, label="Trajetória")
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Lançamento Oblíquo de uma Partícula')
plt.grid(True)
plt.legend()
plt.show()
```

```python
dt = 0.01
fx = 0
fy = -9.8
k_arrasto = 0.3
limites = (0, 20, 0, 15)  # x_min, x_max, y_min, y_max

# Cria a partícula com arrasto e limites
p = Particula(x=0, y=0, vx=12, vy=12, massa=1, limites=limites, k_arrasto=k_arrasto)
# Simulação por tempo máximo ou até velocidade ser muito baixa
tempo_max = 11  # segundos
while p.t < tempo_max and not (abs(p.vx) < 0.1 and abs(p.vy) < 0.1 and p.y <= limites[2]):
    p.newton(fx, fy, dt)

# Gráfico da trajetória
plt.figure(figsize=(8, 5))
plt.plot(p.l_x, p.l_y, label="Trajetória com arrasto e colisão")
plt.xlim(limites[0], limites[1])
plt.ylim(limites[2], limites[3])
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Simulação com Arrasto e Rebatimento nas Paredes')
plt.grid(True)
plt.legend()
plt.show()
```
