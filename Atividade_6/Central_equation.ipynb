```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n = 2  # Use n=3 so that 2n+1 = 7, matching N
k = np.linspace(-np.pi, np.pi, 100, endpoint=False)
G = 2*np.pi*np.arange(-n, n + 1)

def hamiltonian(k_i, alpha, beta, gama): 
    lambda_k = k_i - G
    H = np.eye(n*2 + 1) * lambda_k**2  # Adjusted to use broadcasting correctly
    H += np.eye(n*2 + 1, k=1) * alpha 
    H += np.eye(n*2 + 1, k=-1) * alpha
    H += np.eye(n*2 + 1, k=2) * beta
    H += np.eye(n*2 + 1, k=-2) * beta
    H += np.eye(n*2 + 1, k=3) * gama
    H += np.eye(n*2 + 1, k=-3) * gama

    return H

eigvals = np.zeros((len(k), len(G))) 
for i in range(len(k)):
    H = hamiltonian(k[i], 10, 12, 16)  # This hamiltonian uses N=7, G with n=3
    eigvals[i] = np.linalg.eigvalsh(H)
```

```python
plt.title('Energie Bands')
plt.plot(k, eigvals)
plt.xlabel('k')
plt.ylabel('Energie')
plt.show()
```
