```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n = 5
k = np.linspace(-np.pi, np.pi, 100, endpoint=False)
G = 2*np.pi*np.linspace(-n, n, n) 

def hamiltonian(k_i, alpha, beta, gama): 
    lambda_k = k_i - G 
    H = np.eye(n) * lambda_k**2  # Adjusted to use broadcasting correctly
    H += np.eye(n, k=1) * alpha 
    H += np.eye(n, k=-1) * alpha
    H += np.eye(n, k=2) * beta
    H += np.eye(n, k=-2) * beta
    H += np.eye(n, k=3) * gama
    H += np.eye(n, k=-3) * gama

    return H
```

```python
eigvals = np.zeros((len(k),n)) 
for i in range (len(k)) :
    H = hamiltonian(k[i], 80, 6, 90)
    eigvals[i] = np.linalg.eigvalsh(H)
```

```python
plt.title('Energie Bands')
plt.plot(k, eigvals)
plt.xlabel('k')
plt.ylabel('Energie')
plt.show()
```
